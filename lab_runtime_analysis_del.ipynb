In this notebook we'll analyze the time complexity of the **del** keyword in python upon the inbuilt datatypes - Lists and Dictionaries

## **List**


> First, we'll be analyzing it for Lists. Let's see the time complexity of **del** arr[i], deleting an element at index position i in an array arr







```python
import numpy as np
import random

# We'll be making a 2D array where on the 0th position of the ith cell on the matrix we'll store the index
# value and on the 1st position the time taken to delete the element at that index

import time

index_time_arr = [[]]
for i in range(1, 10**5, 100):
  array = [random.random() for j in range(10**5)]
  begin = time.time()
  del array[i]
  end = time.time()
  index_time_arr.append([i,(end-begin)])

# We'll be plotting the time taken on the y-axis and the index position being deleted on the x-axis
x_axis, y_axis = [], []
for i in range(1,len(index_time_arr)):
  x_axis.append(index_time_arr[i][0])
  y_axis.append(index_time_arr[i][1])



# Now plotting the graph
from matplotlib import pyplot as plt

plt.plot(x_axis, y_axis)

plt.title("Del List[index] Time Complexity")
plt.xlabel("index ->")
plt.ylabel("Time Taken (seconds) ->")

plt.show()
```

By the graph above it can be concluded that **del** has an *O(n)* time complexity when deleting an element with index *i*

```python
# Now let's see if del takes different time in deleting slices of the list

import time

index_time_arr = [[]]
for i in range(100,10**5, 100):
  array = [random.random() for i in range(10**5)]
  begin = time.time()
  del array[:i]
  end = time.time()
  index_time_arr.append([i,(end-begin)])


x_axis, y_axis = [], []
for i in range(1,len(index_time_arr)):
  x_axis.append(index_time_arr[i][0])
  y_axis.append(index_time_arr[i][1])



# Now plotting the index and the time taken in deletion of slice of array

from matplotlib import pyplot as plt

plt.plot(x_axis, y_axis)

plt.title("Del List[:index] Time Complexity")
plt.xlabel("index ->")
plt.ylabel("Time Taken (seconds) ->")

plt.show()
```

We can see that the **del** arr[a:b] operation is *O(n)* in the worst case in a List

Now let's see the time complexity of del keyword when deleting an entire list

```python
# We'll be deleting lists of various sizes
size_time_arr = [[]]
for i in range(100,10**5, 100):
  array = [random.random() for i in range(i)]
  begin = time.time()
  del array
  end = time.time()
  size_time_arr.append([i,(end-begin)])

x_axis, y_axis = [], []
for i in range(1,len(size_time_arr)):
  x_axis.append(size_time_arr[i][0])
  y_axis.append(size_time_arr[i][1])


plt.plot(x_axis, y_axis)

plt.title("Del List Time Complexity")
plt.xlabel("Size of the List ->")
plt.ylabel("Time Taken (seconds) ->")

plt.show()
```

We can see that the **del** arr operation is *O(n)* in the worst case in a List

## **Dictionary**

Let's see if **del** is key position independent. i.e - We need to see if del will taken different time in deleting a key-value pair in a dictionary located at different positions

```python
import random
import time

# Creating a 2D array to store the key position and the time taken to delete the key-value pair at that position
key_del_time_arr = [[]]

for i in range(1,10**5,100):
  # Creating a dictionary
  dict_1 = {j: random.random() for j in range(10**5)}
  a = time.time()
  del dict_1[i]
  b = time.time()
  key_del_time_arr.append([i,(b-a)])

# Now plotting the result

x_axis, y_axis = [],[]

for i in range(1, len(key_del_time_arr)):
  x_axis.append(key_del_time_arr[i][0])
  y_axis.append(key_del_time_arr[i][1])

plt.plot(x_axis, y_axis)

plt.title("Del Dictionary[key] Time Complexity")
plt.xlabel("Key Position in the Dictionary ->")
plt.ylabel("Time Taken (seconds) ->")

plt.show()
```

We can clearly see that **del** is an *O(1)* operation in a dictionary, all the key value pairs are deleted in constant time, irrespective of the key position in the dictionary

Now let's see the time complexity of deleting the entire dictionary using the del operator

```python
# Creating a 2D array to store the key position and time taken to delete the entire dictionary

dict_del_time_arr = [[]]

for i in range(1, 10**5,100):
  # Creating a dictionary
  dict_1 = {j:random.random() for j in range(i)}
  a = time.time()
  del dict_1
  b = time.time()
  dict_del_time_arr.append([i, (b-a)])

# Now plotting the result

x_axis, y_axis = [],[]

for i in range(1, len(dict_del_time_arr)):
  x_axis.append(dict_del_time_arr[i][0])
  y_axis.append(dict_del_time_arr[i][1])


plt.plot(x_axis, y_axis)

plt.title("Del Dictionary Time Complexity")
plt.xlabel("Number of keys in the dictionary ->")
plt.ylabel("Time Taken (seconds) ->")

plt.show()
```

We can clearly see that deleting the dictionary using the **del** keyword is an *O(n)* operation
